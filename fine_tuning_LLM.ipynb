## Exploratory Data Analysis

```python
import pandas as pd
from datasets import Dataset, DatasetDict
from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments

# Load and preprocess the data
df = pd.read_csv('Reviews.csv')
df.head(10)
```

##

```python
df = df.dropna(subset=['Score', 'Text'])
df = df.drop_duplicates(subset=['Text'])
```

```python


df['Sentiment'] = df['Score'].apply(lambda x: 1 if x > 3 else 0)
df['Text'] = df['Text'].str.lower().str.replace(r'[^\w\s]+', '', regex=True)

pd.set_option('display.max_colwidth', None)
keywords = ["shrunk", "inconsistency in sizing", "size mismatch", "incorrect size", "wrong size",
            "sizing issue", "fit problem", "size discrepancy", "not true to size", "size error",
            "size inconsistency", "wrong fit", "size problem", "size variation",
            "incorrect measurements", "size deviation", "size not as described"]

pattern = '|'.join(keywords)
df_keyword= df[df['Text'].str.contains(pattern, case=False, na=False)]

combined_df = pd.concat([df, df_keyword]).drop_duplicates().reset_index(drop=True)

# Convert to Hugging Face Dataset
dataset = Dataset.from_pandas(combined_df[['Text', 'Sentiment']].head(4000))

# Split the dataset into train and test
dataset = dataset.train_test_split(test_size=0.2)
```

```python
# Load tokenizer
tokenizer = BertTokenizer.from_pretrained('bert-base-uncased')

# Tokenize the dataset
def tokenize_function(examples):
    return tokenizer(examples['Text'], padding='max_length', truncation=True, max_length=512)

tokenized_train_dataset = dataset['train'].map(tokenize_function, batched=True)
tokenized_test_dataset =  dataset['test'].map(tokenize_function, batched=True)

# Rename columns and set format
tokenized_train_dataset = tokenized_train_dataset.rename_column('Sentiment', 'labels')
tokenized_train_dataset = tokenized_train_dataset.remove_columns(['Text'])
tokenized_train_dataset.set_format('torch')

tokenized_test_dataset = tokenized_test_dataset.rename_column('Sentiment', 'labels')
tokenized_test_dataset = tokenized_test_dataset.remove_columns(['Text'])
tokenized_test_dataset.set_format('torch')
```

```python
# Assuming tokenized_train_dataset and tokenized_test_dataset are Dataset objects
tokenized_datasets = DatasetDict({
    'train': tokenized_train_dataset,
    'test': tokenized_test_dataset
})

# Save the datasets
tokenized_datasets.save_to_disk('tokenized_datasets')
```

## prediction function

```python
# Load the fine-tuned model and tokenizer for inference
model = BertForSequenceClassification.from_pretrained("Amazon_tuned_model")
tokenizer = BertTokenizer.from_pretrained("Amazon_tuned_model")
# Function to make predictions on new text data
def predict(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    model.eval()
    with torch.no_grad():
        outputs = model(**inputs)
    logits = outputs.logits
    predicted_class = torch.argmax(logits, dim=-1).item()
    label = "positive" if predicted_class == 1 else "negative"
    return label

# Test the prediction function
text1 = "I'd buy this product next time."
text2 = "The product size is slightly larger than usual."

print(f"Text: {text1}\nPrediction: {predict(text1)}\n")
print(f"Text: {text2}\nPrediction: {predict(text2)}\n")
```
